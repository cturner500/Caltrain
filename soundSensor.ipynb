```python
#Now, let's import our cleansed sensor data (with Caltrain schedule appended)
```

```python
def getColumn(filename, column):
    results = csv.reader(open(filename, 'rU'), delimiter=",")
    #results = csv.reader(open(filename), dialect=csv.excel_tab)
    return [result[column] for result in results]
#trains = genfromtxt("soundSensorWithTrains.csv", delimiter=',')

epoch = getColumn("soundSensorWithTrains.csv",0)
DateTime = getColumn("soundSensorWithTrains.csv",1)
Date = getColumn("soundSensorWithTrains.csv",2)
Time = getColumn("soundSensorWithTrains.csv",3)
SoundLevel = getColumn("soundSensorWithTrains.csv",4)
NB_PaloAlto = getColumn("soundSensorWithTrains.csv",5)
SB_CalAve = getColumn("soundSensorWithTrains.csv",6)
dB = getColumn("soundSensorWithTrains.csv",7)
```

```python
fltSoundLevel = [float(i) for i in SoundLevel]
fltNB_PaloAlto = [float(i)*4 for i in NB_PaloAlto]
fltSB_CalAve = [float(i)*4 for i in SB_CalAve]
fltDateTime = [float(i) for i in DateTime]
fltdB = [float(i) for i in dB]
fltepoch = [float(i) for i in epoch]
strTime = [time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(i)) for i in fltepoch]
 
```

```python
mean(fltNB_PaloAlto)
```

```python
mean(fltSB_CalAve)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.set_xticks(fltepoch[0::10000])
ax.set_xticklabels(strTime[0::10000], rotation=90)
ax.set_xlabel('Local Time (PST)')
ax.set_ylabel('Microphone Output (VDC)')
plot(epoch, SoundLevel, 'b-', epoch, fltNB_PaloAlto, 'r-', epoch, fltSB_CalAve, 'g-')
```

```python
#Now let's take a look at this as decibels vs. voltage:
```

```python
dbPA = [i*50 for i in fltNB_PaloAlto]
dbCA = [i*50 for i in fltSB_CalAve]
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ylim([40,90.0])
#ax.set_xticklabels(strTime, rotation=90)
ax.set_xticks(fltepoch[0::10000])
ax.set_xticklabels(strTime[0::10000], rotation=90)
ax.set_xlabel('Local Time (PST)')
ax.set_ylabel('Sound Level (dB)')
plot(epoch, fltdB, 'b-', epoch, dbPA, 'r-', epoch, dbCA, 'g-')
```

```python
#Now let's look at just a subset that looks interesting and add time labels:
fig, ax = plt.subplots(figsize=(20,10))
ylim([40,100.0])
ax.set_xticks(fltepoch[30000::100])
ax.set_xticklabels(strTime[30000::100], rotation=90)
#ax.set_xticklabels(strTime[30000:31000][0::100],rotation=90)
ax.set_xlabel('Local Time (PST)')
ax.set_ylabel('Sound Level (dB)')
ax.fill_between(fltepoch[30000:31000],dbPA[30000:31000], 0,facecolor='red')
ax.fill_between(fltepoch[30000:31000],dbCA[30000:31000], 0,facecolor='green')
plot(fltepoch[30000:31000], fltdB[30000:31000], 'b-', epoch[30000:31000], dbPA[30000:31000], 'r-', epoch[30000:31000], dbCA[30000:31000], 'g-')
```

```python
#Let's zoom in even further and look at a single stoplight change:
fig, ax = plt.subplots(figsize=(20,10))
ylim([40,90.0])
ax.set_xticks(arange(fltepoch[30000], fltepoch[30100], 10))
ax.set_xticklabels(strTime[30000::10], rotation=90)
ax.set_xlabel('Local Time (PST)')
ax.set_ylabel('Sound Level (dB)')
plot(epoch[30000:30100], fltdB[30000:30100], 'b-', epoch[30000:30100], dbPA[30000:30100], 'r-', epoch[30000:30100], dbCA[30000:30100], 'g-')
```
